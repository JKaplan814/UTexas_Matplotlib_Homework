```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinic_df = pd.read_csv(clinical_trial_data_to_load)
# Combine the data into a single dataset
df = pd.merge(clinic_df,mouse_df,on='Mouse ID',how='outer')

# Display the data table for preview
df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped = df.groupby(['Drug','Timepoint'])
mean_tumor = grouped['Tumor Volume (mm3)'].mean()

# Convert to DataFrame
tumor_df = pd.DataFrame({
    'Tumor Volume (mm3)':mean_tumor
}).reset_index()

# Preview DataFrame
tumor_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
error = grouped['Tumor Volume (mm3)'].sem()

# Convert to DataFrame
error_df = pd.DataFrame({
    'Tumor Volume (mm3) SEM':error
}).reset_index()

# Preview DataFrame

error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_sizes = tumor_df.pivot(index = 'Timepoint',columns='Drug',values='Tumor Volume (mm3)')

# Preview that Reformatting worked
tumor_sizes
```

```python
# Generate the Plot (with Error Bars)
errors = error_df.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3) SEM')
tumor_sizes.loc[:,['Capomulin','Infubinol','Ketapril','Placebo']].plot(kind='line',grid=True,marker='o',yerr=errors.loc[:,['Capomulin','Infubinol','Ketapril','Placebo']],linewidth=.7,elinewidth=.7,figsize=(10,5))
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Tumor Response to Treatment')

# Save the Figure

plt.savefig('Tumor Response to Treatment.png')
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint
mean_met = grouped['Metastatic Sites'].mean()

# Convert to DataFrame
met_df = pd.DataFrame({
    'Metastatic Sites':mean_met
}).reset_index()

# Preview DataFrame
met_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_error = grouped['Metastatic Sites'].sem()

# Convert to DataFrame
met_error_df = pd.DataFrame({
    'Metastatic Sites SEM':met_error
}).reset_index()

# Preview DataFrame
met_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_sites_df = met_df.reset_index()
met_sites = met_sites_df.pivot(index = 'Timepoint',columns='Drug',values='Metastatic Sites')

# Preview that Reformatting worked
met_sites.head()
```

```python
errors = met_error_df.pivot(index='Timepoint',columns='Drug',values='Metastatic Sites SEM')
met_sites.loc[:,['Capomulin','Infubinol','Ketapril','Placebo']].plot(kind='line',grid=True,marker='o',yerr=errors.loc[:,['Capomulin','Infubinol','Ketapril','Placebo']],linewidth=.7,elinewidth=.7,figsize=(10,5))

plt.ylabel('Metastatic Sites')
plt.title('Metastatic Spread During Treatment')


#Save the Figure
plt.savefig('Metastatic Spread During Treatment.png')
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count = grouped['Mouse ID'].count()

# Convert to DataFrame
mice_df = pd.DataFrame({
    'Mouse Count':mice_count
}).reset_index()

# Preview DataFrame
mice_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_pivot = mice_df.pivot(index='Timepoint',columns='Drug',values='Mouse Count')
# Preview the Data Frame
mice_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
mice_survival = mice_pivot.divide(mice_pivot.ix[0])
mice_survival.loc[:,['Capomulin','Infubinol','Ketapril','Placebo']].plot(kind='line',grid=True,marker='o',linewidth=.7)
plt.title('Survival During Treatment')
plt.ylabel('Survival Rate (%)')
plt.xlabel('Time (Days)')

# Save the Figure
plt.savefig('Survival During Treatment.png')
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
average_change = ((tumor_sizes.tail(1)/ tumor_sizes.ix[0] -1)*100).squeeze()

# Display the data to confirm
average_change
```

```python
# Store all Relevant Percent Changes into a Tuple
key_drugs = average_change.loc[['Capomulin','Infubinol','Ketapril','Placebo']]
drug_names = key_drugs.index

# Splice the data between passing and failing drugs
increased_growth = key_drugs.loc[key_drugs > 0]
decreased_growth = key_drugs.loc[key_drugs < 0]
x_axis = np.arange(len(increased_growth)+len(decreased_growth))
#x_inc = len(increased_growth)
#x_dec = len(decreased_growth)

fig, ax = plt.subplots()
# Orient widths. Add labels, tick marks, etc. 
rects = summary_plot = plt.bar(x=x_axis,height=key_drugs,width=-1,color='red',align='edge')
plt.xlim(-1.25,len(x_axis)-.75)
plt.xticks(x_axis,drug_names)
plt.yticks(np.arange(-20,70,20))
plt.xlabel('Drug')
plt.ylabel('% Tumor Volume Change')
plt.title('Tumor Change Over 45 Day Treatment')
summary_plot[0].set_color('g')
plt.grid()

# Use functions to label the percentages of changes
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., .5*height,
                '%d' % int(height) + '%',
                ha='center', va='bottom')

# Call functions to implement the function calls
autolabel(rects)

# Save the Figure
plt.savefig('Tumor Change Over Treatment.png')
# Show the Figure
plt.show()
```
